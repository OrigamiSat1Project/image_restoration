```python
import pandas as pd
import numpy as np
import os
import codecs
import csv
```

```python
file_name = "test.log"
```

## Reading file

```python
dataSize = 36
init_del = 4
col_names = ['{0:d}'.format(i) for i in range(dataSize + init_del)]
```

```python
data=csv.reader(file_name)
print(data)
```

```python
with  codecs.open(file_name, 'r', 'utf-8', 'ignore') as f:
    df = pd.read_csv(f, header=None, sep=' ', names=col_names)
# namesで列を指定することで、列が1つだけになる事を防ぐ
# ignoreでエラーが出ても読み続ける

init_len = len(df)

df = df.drop(range(0,init_len,6))
df = df.drop(range(1,init_len,6))
# df = df.drop(range(2,6*init_del,6))
# df = df.drop(range(3,6*init_del,6))
df = df.drop(range(4,init_len,6))
df = df.drop(range(5,init_len,6))
# # axisの初期条件は行指定。rangeで0からinit_lenまで6飛ばしで消す

df = df.drop(['0','1','2','3'], axis=1)
# 列はnames=col_namesで作ったため、文字コードの'0'とかになる
# df

# df.applymap(type)
# Databaseの型を表示する

# df = df.astype(str)
# > を削除するためにapplyを使いたく、そのために一度str型にしておく
# ここでstr型にすることによってNaNがnanとなり、扱いが面倒になった

# ret = df.apply(lambda d: d.str.contains('>', na=False))
# df = df.mask(ret)
# df
# >をマスクしてNaNにする

# df.reindex(range(32), axis=1)
# df.reindex(range(len(df)), axis=0)
# indexを綺麗にする。でも、あまりうまくいかない

# df = df.apply(lambda s: pd.Series(s.dropna().tolist()), axis=1)
# NaNを消してリストにすることで詰めている 

df = df.replace('>', np.nan)
# >をNaNにすることで、あとで消しやすくする

def append_2lines(line1, line2):
    tmp_series = pd.Series(line1)
    tmp_series = tmp_series.append(line2)
    tmp_series = tmp_series.dropna()
    return tmp_series

df
```

```python
resize_len = int(len(df)/2)

init_list = append_2lines(df.iloc[0], df.iloc[1]).tolist()
new_df = pd.DataFrame([init_list])
# 一度リスト型にすることで、columnsの名前をリセットでき、それをSeriesに戻してから連結することで、列番号が自然につく

for i in range(1,resize_len):
    tmp_list = append_2lines(df.iloc[2*i], df.iloc[2*i+1]).tolist()
    tmp_s = pd.Series(tmp_list, name=i)
    new_df = new_df.append(tmp_s)
new_df
```

## CSVへ出力

```python
new_df.to_csv("output.csv", index=False, header=False)
```
